# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("C:/Users/dax98/Downloads/boston housing.xls")
```

```python
df
```

```python
y=df['MV']
```

```python
y
```

```python
X=df.drop("MV",axis=1)
```

```python
X
```

```python
lm = LinearRegression()
```

```python
lm.fit(X,y)
```

```python
lm.predict(X)
```

```python
print(lm.intercept_,lm.coef_)
```

```python
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
print(predictions)
```

```python
model.summary()
```

```python
Z=df['RM']
```

```python
model = sm.OLS(y, Z).fit()
```

```python
model.summary()
```

```python
z=df['NOX']
z
```

```python
model = sm.OLS(y, z).fit()
```
